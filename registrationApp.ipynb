```python
from tkinter import *
from customtkinter import *
from tkinter import messagebox
import ast

# Initialize the main window
window = Tk()
window.title("SignUp/Login")
window.geometry('925x500+300+200')
window.configure(bg='#fff')
window.resizable(False, False)

# Placeholder functions
def add_placeholder(entry, placeholder_text):
    """Add placeholder text to an Entry widget."""
    entry.insert(0, placeholder_text)
    entry.config(fg='grey')
    entry.bind('<FocusIn>', lambda e: clear_placeholder(e, placeholder_text))
    entry.bind('<FocusOut>', lambda e: set_placeholder(e, placeholder_text))

def clear_placeholder(event, placeholder_text):
    """Clear placeholder text on focus in."""
    if event.widget.get() == placeholder_text:
        event.widget.delete(0, END)
        event.widget.config(fg='black')

def set_placeholder(event, placeholder_text):
    """Set placeholder text on focus out if the field is empty."""
    if not event.widget.get():
        event.widget.insert(0, placeholder_text)
        event.widget.config(fg='grey')

# Toggle password visibility for sign-up password fields
def toggle_password(entry, button):
    if entry.cget('show') == '*':
        entry.config(show='')
        button.config(text='üëÅÔ∏è')  # Change icon to indicate visible
    else:
        entry.config(show='*')
        button.config(text='üôà')  # Change icon to indicate hidden

# Toggle password visibility for login password field
def toggle_login_password():
    toggle_password(login_code, toggle_login_btn)

# Function to switch to login frame
def switch_to_login():
    signup_frame.pack_forget()
    login_frame.pack(side="right", padx=(50, 100))

# Function to switch to signup frame
def switch_to_signup():
    login_frame.pack_forget()
    signup_frame.pack(expand=False, side="right", padx=(50, 100))

# Function to handle the sign-up process
def signup():
    username = user.get()
    password = code.get()
    confirm_password = confirm_code.get()

    if username in ("Username", "") or password in ("Password", "") or confirm_password in ("Confirm Password", ""):
        messagebox.showerror('Error', 'All fields are required!')
        return

    if password != confirm_password:
        messagebox.showerror('Error', "Both passwords should match")
        return

    try:
        with open('datasheet.txt', 'r+') as file:
            data = file.read()
            users = ast.literal_eval(data) if data else {}

        if username in users:
            messagebox.showerror('Error', 'Username already exists')
        else:
            users[username] = password
            with open('datasheet.txt', 'w') as file:
                file.write(str(users))
            messagebox.showinfo('Success', 'Sign up successful!')
            switch_to_login()
    except FileNotFoundError:
        with open('datasheet.txt', 'w') as file:
            file.write(str({username: password}))
        messagebox.showinfo('Success', 'Sign up successful!')
        switch_to_login()

# Function to handle the login process
def login():
    username = login_user.get()
    password = login_code.get()

    try:
        with open('datasheet.txt', 'r') as file:
            users = ast.literal_eval(file.read())

        if users.get(username) == password:
            messagebox.showinfo('Success', 'Login successful!')
        else:
            messagebox.showerror('Error', 'Invalid credentials or sign up required')
    except FileNotFoundError:
        messagebox.showerror('Error', 'No user data found, please sign up first')

# Load and place the image on the left side
img = PhotoImage(file='loginImage.png')
Label(window, image=img, border=0, bg='white').place(x=100, y=90)

# Sign-Up Frame Setup
signup_frame = Frame(window, width=350, height=390, bg='white')
heading1 = Label(signup_frame, text='Sign Up', fg='#57a1f8', bg='white', font=('Microsoft Yahei UI Light', 23, 'bold'))
heading1.place(x=100, y=5)

# Username Entry with Placeholder
user = Entry(signup_frame, width=35, fg='black', border=2, bg='white', font=('Microsoft Yahei UI Light', 11))
user.place(x=30, y=80)
add_placeholder(user, 'Username')

# Password Entry with Placeholder
code = Entry(signup_frame, width=27, fg='black', border=2, bg='white', font=('Microsoft Yahei UI Light', 11), show='*')
code.place(x=30, y=150)
add_placeholder(code, 'Password')

# Eye button for password field
toggle_signup_btn = Button(signup_frame, text='üëÅÔ∏è', command=lambda: toggle_password(code, toggle_signup_btn), border=0, bg='white', cursor='hand2')
toggle_signup_btn.place(x=285, y=150)

# Confirm Password Entry with Placeholder
confirm_code = Entry(signup_frame, width=27, fg='black', border=2, bg='white', font=('Microsoft Yahei UI Light', 11), show='*')
confirm_code.place(x=30, y=220)
add_placeholder(confirm_code, 'Confirm Password')

# Eye button for confirm password field
toggle_confirm_btn = Button(signup_frame, text='üëÅÔ∏è', command=lambda: toggle_password(confirm_code, toggle_confirm_btn), border=0, bg='white', cursor='hand2')
toggle_confirm_btn.place(x=285, y=220)

# Sign Up Button
CTkButton(signup_frame, text='Sign up', command=signup).place(x=35, y=280)

# Switch to Login Label
Label(signup_frame, text='Already have an account?', fg='black', bg='white', font=('Microsoft YaHei UI Light', 9)).place(x=90, y=340)
signin_btn = Button(signup_frame, text='Log in', border=0, bg='white', cursor='hand2', fg='#57a1f8', command=switch_to_login)
signin_btn.place(x=240, y=341)

# Login Frame Setup
login_frame = Frame(window, width=350, height=390, bg='white')
heading2 = Label(login_frame, text='Log in', fg='#57a1f8', bg='white', font=('Microsoft Yahei UI Light', 23, 'bold'))
heading2.place(x=100, y=5)

# Login Username Entry with Placeholder
login_user = Entry(login_frame, width=35, fg='black', border=2, bg='white', font=('Microsoft Yahei UI Light', 11))
login_user.place(x=30, y=80)
add_placeholder(login_user, 'Username')

# Login Password Entry with Placeholder
login_code = Entry(login_frame, width=27, fg='black', border=2, bg='white', font=('Microsoft Yahei UI Light', 11), show='*')
login_code.place(x=30, y=150)
add_placeholder(login_code, 'Password')

# Eye button for login password field
toggle_login_btn = Button(login_frame, text='üëÅÔ∏è', command=toggle_login_password, border=0, bg='white', cursor='hand2')
toggle_login_btn.place(x=285, y=150)

# Login Button
CTkButton(login_frame, text='Log in', command=login).place(x=35, y=220)

# Switch to Signup Label
Label(login_frame, text="Don't have an account?", fg='black', bg='white', font=('Microsoft YaHei UI Light', 9)).place(x=90, y=280)
signup_btn = Button(login_frame, text='Sign up', border=0, bg='white', cursor='hand2', fg='#57a1f8', command=switch_to_signup)
signup_btn.place(x=230, y=281)

# Start with Signup Frame
switch_to_signup()

# Run the application
window.mainloop()
```
